```python
import pandas as pd
import copy
```

```python
matching = pd.read_csv('poms-bix.scoring', sep = '\t', names = ['index','score','dimension','word'])
del matching['index']
matching.index = matching.word
del matching['word']

to_match = pd.read_csv('POMS_ngrams.txt', sep = '\t', names = ['ref','word','score'])
to_match.ref = to_match.ref.str.lower()
to_match.word
```

```python
matching.head()
```

```python
results = []
used = []

dimensions = matching.dimension.unique().tolist()

for word in matching.index.unique():
    used.append(word)
    score = matching[matching.index == word].score[0]
    dimension =  matching[matching.index == word].dimension[0]
    
    r = {}
    r[word] = {'composed/anxious':0,'agreeable/hostile':0,'elated/depressed':0,'confident/unsure':0,
              'clearheaded/confused':0,'energetic/tired':0}
    r[word][dimension] += score
            
    results.append(r)
```

```python
for word in to_match.word:
    
    if word not in used:
    
        used.append(word)
        scores = to_match[to_match.word == word].score.tolist()
        dimensions =  to_match[to_match.word == word].ref.apply(lambda x: matching[matching.index == x].dimension[0]).tolist()
        multipliers = to_match[to_match.word == word].ref.apply(lambda x: matching[matching.index == x].score[0]).tolist()
        r = {}

        r[word] = {'composed/anxious':0,'agreeable/hostile':0,'elated/depressed':0,'confident/unsure':0,
                  'clearheaded/confused':0,'energetic/tired':0}
        
        for x in range(len(scores)):
            r[word][dimensions[x]] += scores[x] * multipliers[x]
        results.append(r)
```

```python
df_results = []
for x in results:
    word = list(x.keys())[0]
    r = [word, x[word]['composed/anxious'], x[word]['agreeable/hostile'], x[word]['elated/depressed'], x[word]['confident/unsure']
        , x[word]['clearheaded/confused'], x[word]['energetic/tired']]
    df_results.append(r)
```

```python
columns = ['Word', 'composed/anxious','agreeable/hostile','elated/depressed','confident/unsure','clearheaded/confused','energetic/tired']
df = pd.DataFrame(df_results, columns = columns)
df.index = df['Word']
del df['Word']

df
```

```python
df.to_csv('POMS.csv', sep = '\t')
```
